```python
#Create a scatter plot to show surface area vs charging station
# ax=plt.axes()
# ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
# ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['Land_Surface_Area'], Rates_Population_df["Public_Charging_Station"], c = Rates_Population_df["Public_Charging_Station"]) 
clb = plt.colorbar()
plt.xlabel("Suface Area")
plt.ylabel("Public Charging Station")
plt.title("Surface Area Vs Charging Station")
clb.set_label("Public_Charging_Station")

plt.show()
plt.savefig("scatterplot.png")

#Analysis
# There appears to be no correlation between the Surface area public charging stations
```

```python
#Number of electric vehicles by county vs Electric cost

ax=plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['Electric_ Cost_KWhr'], Rates_Population_df['EVs_Per_County']) 
# clb = plt.colorbar()
plt.xlabel("Electrical Cost KW/Hour")
plt.ylabel("Electrical Vehicle Count ")
plt.title("Electrical Cost vs Electrical Vehicle Count")
# clb.set_label("Public Charging Station ")
# figsize
plt.show()
```

```python
#number of Electrical vehicle vs population 

plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['EVs_Per_County'], Rates_Population_df['Population']) 
plt.xlabel("Population by County")
plt.ylabel("Number of Electrical Vehicle")
plt.title("Electrical Vehicle Vs Population")
plt.show()
```
