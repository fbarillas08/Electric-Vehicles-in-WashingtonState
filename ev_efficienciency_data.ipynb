```python
#Import Dependencies
import matplotlib.ticker as ticker
import pandas as pd
import pprint
import numpy as np
import matplotlib.pyplot as plt
import re 
import scipy as sp
```

```python
#Import cvs
# Washington_Data = "Wash_State_EV_PHEV_Data_Expanded.csv"

Rates_Population = "Wash_State_County_Demo_Data.csv"

#read cvs

Rates_Population_df = pd.read_csv(Rates_Population)

Rates_Population_df.head()
```

```python
Rates_Population_df.columns
```

```python
#Create a bar graph to show charging stations by county



sorted_charging_station = Rates_Population_df.sort_values('Public Charging Station ')


# Rates_Population_df[['County ','Public Charging Station ']].set_index('County ').plot(kind='bar')
ax=plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
plt.xticks(rotation = 90)
plt.bar('County ', 'Public Charging Station ', data = sorted_charging_station)
plt.xlabel("County", size=15)
plt.ylabel("Number of Public Charging Stations", size = 15)
plt.title("Number of Charging Stations per County", size = 15)

#turn into two axis with ascending order 
#second axis number of cars 
```

```python
#Create a scatter plot to show charging stations vs surface area
ax=plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['Land Surface Area'], Rates_Population_df["Public Charging Station "], c = Rates_Population_df["Public Charging Station "]) 
clb = plt.colorbar()
plt.xlabel("Suface Area")
plt.ylabel("Public Charging Station ")
plt.title("Surface Area Vs Charging Station")
clb.set_label("Public Charging Station ")

plt.show()
```

```python
#Number of electric vehicles by county vs Electric cost
ax=plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['Electric_ Cost_KWhr'], Rates_Population_df['EVs_Per_County']) 
# clb = plt.colorbar()
plt.xlabel("Electrical Cost KW/Hour")
plt.ylabel("Electrical Vehicle Count ")
plt.title("Electrical Cost vs Electrical Vehicle Count")
# clb.set_label("Public Charging Station ")
# figsize
plt.show()
```

```python
#number of Electrical vehicle vs population 
ax=plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['Population'], Rates_Population_df['EVs_Per_County'] ) 
# clb = plt.colorbar()
# plt.rc('grid', linestyle="-", color='black')
plt.xlabel("Population by County")
plt.ylabel("Number of Electrical Vehicle")
plt.title("Electrical Vehicle Vs Population")
# clb.set_label("Evs per county ")
# figsize
plt.show()
```

```python
#number of Electrical vehicle number vs charging stations

ax=plt.gca()

plt.scatter(Rates_Population_df["Public Charging Station "], Rates_Population_df['EVs_Per_County']) 
# clb = plt.colorbar()


plt.xlabel("Number of Charging Stations")
plt.ylabel("Electrical Vehicle Count ")
plt.title("Electrical Vehicle vs Charging Stations")

# clb.set_label("Public Charging Station ")
# grid=True
# figsize
# plt.show()
plt.savefig("scatter.png")

```
