```python
#Import Dependencies
import matplotlib.ticker as ticker
import pandas as pd
import pprint
import numpy as np
import matplotlib.pyplot as plt
import re 
import scipy as sp
```

```python
#Import cvs

Rates_Population = "Wash_State_County_Demo_Data.csv"

#read cvs

Rates_Population_df = pd.read_csv(Rates_Population)

Rates_Population_df.head()

# Rates_Population_df.dtypes
```

```python
Rates_Population_df.columns
```

```python
#Create a bar graph to show charging stations by county


Rates_Population_df[['County', 'Public_Charging_Station']].set_index('County').plot(kind='bar')
plt.ylabel("Number of Public Charging Stations")
plt.title("Charging Station by County")
plt.grid()
plt.savefig("barplot.png")
```

```python
#Create a scatter plot to show surface area vs charging station
# ax=plt.axes()
# ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
# ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['Land_Surface_Area'], Rates_Population_df["Public_Charging_Station"], c = Rates_Population_df["Public_Charging_Station"]) 
clb = plt.colorbar()
plt.xlabel("Suface Area")
plt.ylabel("Public Charging Station")
plt.title("Surface Area Vs Charging Station")
clb.set_label("Public_Charging_Station")

plt.show()
plt.savefig("scatterplot.png")

#Analysis
# There appears to be no correlation between the Surface area public charging stations
```

```python
#Number of electric vehicles by county vs Electric cost

ax=plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['Electric_Cost_KWhr'], Rates_Population_df['EVs_Per_County']) 
# clb = plt.colorbar()
plt.xlabel("Electrical Cost KW/Hour")
plt.ylabel("Electrical Vehicle Count ")
plt.title("Electrical Cost vs Electrical Vehicle Count")
# clb.set_label("Public Charging Station ")
# figsize
plt.show()
```

```python
#number of Electrical vehicle vs population 

plt.xticks(rotation = 90)
plt.scatter(Rates_Population_df['EVs_Per_County'], Rates_Population_df['Population']) 
plt.xlabel("Population by County")
plt.ylabel("Number of Electrical Vehicle")
plt.title("Electrical Vehicle Vs Population")
plt.show()
```

```python
#number of Electrical vehicle number vs charging stations

ax=plt.gca()

plt.scatter(Rates_Population_df["Public_Charging_Station"], Rates_Population_df['EVs_Per_County']) 
# clb = plt.colorbar()


plt.xlabel("Number of Charging Stations")
plt.ylabel("Electrical Vehicle Count ")
plt.title("Electrical Vehicle vs Charging Stations")

# clb.set_label("Public Charging Station ")
# grid=True
# figsize
# plt.show()
# plt.savefig("scatter.png")

```

```python
#DataFrame with top FIve 

final_df = Rates_Population_df.sort_values(by=['Public_Charging_Station'], ascending=False).head(5)
final_df


# Analysis : Population is not correlated to the number of public charging stations. 
#In fact, Spokane county has a higher population but significantly less public charging stations than Stevens County. 
```

```python

# Top Five Counties With Public Charging Stations
labels = ['King', 'Spokane', 'Pierce', 'Snohomish','Clark']
sizes = [47.5, 30.0, 10.0, 7.0, 5.0]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','red']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90,)
plt.legend(patches, labels, loc="best")
plt.title("Counties With Highest Public EV Charging Stations")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
